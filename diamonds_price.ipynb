# Seleksi fitur

```python
features = ['carat','depth','table','x','y','z']
x = df[features]
y = df['price']
x.shape, y.shape 
```

# Split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# Membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
Ir = LinearRegression()
Ir.fit(x_train,y_train)
pred = Ir.predict(x_test)
```

```python
score = Ir.score(x_test, y_test)
print('akurasi model regresi linier = ', score)
```

# Membuat inputan model regresi linier

```python
#carat=0.23, depth=21.5, table=55.0, x=3.95, y=3.98 z=2.43
input_data = np.array([[0.23,21.5,55.0,3.95,3.98,2.43]])

prediction = Ir.predict(input_data)
print('Harga Berlian dalam EUR :', prediction)
```
